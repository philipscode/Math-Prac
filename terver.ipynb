```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform

%matplotlib inline
plt.style.use('classic')
```

```python
dist = uniform(0, 100)
sample = dist.rvs(100)
```

```python
plt.figure(figsize=(12, 8))
plt.hist(sample)
plt.title('Sample Histogram')
plt.xlabel('x')
plt.ylabel('Frequency')
plt.show()
```

```python
print('Sample median: ', np.median(sample))
print('Sample range: ', np.max(sample) - np.min(sample))
print('Sample variance: ', np.var(sample))
print('Sample mean: ', np.mean(sample))
```

```python
def plot_estimator(estimator, low, high, title, ylabel):
    plt.figure(figsize=(12, 8))
    x, y = estimator(low, high)
    y_mean = [np.mean(y)] * np.size(y)
    plt.plot(x, y)
    plt.plot(x, y_mean, 'r', linewidth=2)
    plt.title(title)
    plt.xlabel('size')
    plt.ylabel(ylabel)
    plt.show()
```

```python
def median_estimator(low, high):
    sample_size = []
    medians = []
    dist = uniform(low, high)
    for size in range(10, 10000, 10):
        sample = dist.rvs(size)
        sample_size.append(np.size(sample))
        medians.append(np.median(sample))
    return sample_size, medians
```

```python
plot_estimator(median_estimator, low=0, high=100, title='Medias-Size', ylabel='median')
```

```python
def range_estimator(low, high):
    sample_size = []
    ranges = []
    dist = uniform(low, high)
    for size in range(10, 10000, 10):
        sample = dist.rvs(size)
        sample_size.append(np.size(sample))
        ranges.append(np.max(sample) - np.min(sample))
    return sample_size, ranges
```

```python
plot_estimator(range_estimator, low=0, high=100, title='Range-Size', ylabel='range')
```

```python
def mean_estimator(low, high):
    sample_size = []
    means = []
    dist = uniform(low, high)
    for size in range(10, 10000, 10):
        sample = dist.rvs(size)
        sample_size.append(np.size(sample))
        means.append(np.mean(sample))
    return sample_size, means
```

```python
plot_estimator(mean_estimator, low=0, high=100, title='Mean-Size', ylabel='mean')
```

```python
def variance_estimator(low, high):
    sample_size = []
    variances = []
    dist = uniform(low, high)
    for size in range(10, 10000, 10):
        sample = dist.rvs(size)
        sample_size.append(np.size(sample))
        variances.append(np.var(sample))
    return sample_size, variances
```

```python
plot_estimator(variance_estimator, low=0, high=100, title='Variance-Size', ylabel='variance')
```

```python
def plot_deviation(deviation, low, high, title):
    plt.figure(figsize=(12, 8))
    plt.plot(*deviation(low, high))
    plt.title(title)
    plt.xlabel('size')
    plt.ylabel('deviation')
    plt.show()
```

```python
def median_deviation(low, high):
    dist = uniform(low, high)
    dist_median = dist.median()
    sizes = []
    median_deviations = []
    print('Distribution median: ', dist_median)
    for size in range(10, 10000, 10):
        sample = dist.rvs(size)
        sample_median = np.median(sample)
        sizes.append(size)
        median_deviations.append(np.abs(dist_median - sample_median))
    return sizes, median_deviations
```

```python
plot_deviation(median_deviation, low=0, high=100, title='Median Deviation')
```

```python
def range_deviation(low, high):
    dist = uniform(low, high)
    dist_range = high - low
    sizes = []
    range_deviations = []
    print('Distribution range: ', dist_range)
    for size in range(10, 10000, 10):
        sample = dist.rvs(size)
        sample_range = np.max(sample) - np.min(sample)
        sizes.append(size)
        range_deviations.append(np.abs(dist_range - sample_range))
    return sizes, range_deviations
```

```python
plot_deviation(range_deviation, low=0, high=100, title='Range Deviation')
```

```python
def mean_deviation(low, high):
    dist = uniform(low, high)
    dist_mean = dist.mean()
    sizes = []
    mean_deviations = []
    print('Distribution mean: ', dist_mean)
    for size in range(10, 10000, 10):
        sample = dist.rvs(size)
        sample_mean = np.mean(sample)
        sizes.append(size)
        mean_deviations.append(np.abs(dist_mean - sample_mean))
    return sizes, mean_deviations
```

```python
plot_deviation(mean_deviation, low=0, high=100, title='Mean Deviation')
```

```python
def variance_deviation(low, high):
    dist = uniform(low, high)
    dist_variance = dist.var()
    sizes = []
    variance_deviations = []
    print('Distribution variance: ', dist_variance)
    for size in range(10, 10000, 10):
        sample = dist.rvs(size)
        sample_variance = np.var(sample)
        sizes.append(size)
        variance_deviations.append(np.abs(dist_variance - sample_variance))
    return sizes, variance_deviations
```

```python
plot_deviation(variance_deviation, low=0, high=100, title='Variance Deviation')
```
